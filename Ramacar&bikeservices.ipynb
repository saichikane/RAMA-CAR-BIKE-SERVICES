Dealing with null values & we cannot remove duplicate here because data can repeated if one person book ride multiple times

change null values with ("  ") space because if we put 0 then there was some columns which allready have value of 0 inside them , same with 'NO' As well 50/70 % data is null so we cannot remove it direct in that case we can,t get any data there , just need to filtered data when we making dashboard power bi

```python
#REMOVING NULL VALUES COLUMN WISE

final_df["Customer_Rating"].fillna( 00 , inplace=True)
final_df["Driver_Ratings"].fillna(00, inplace=True)
final_df['Payment_Method'].fillna("NO", inplace=True)

final_df["Incomplete_Rides_Reason"].fillna("NO", inplace=True)
final_df["Incomplete_Rides"].fillna( "NAN" , inplace=True)
final_df["Canceled_Rides_by_Driver"].fillna("NO", inplace=True)
final_df['Canceled_Rides_by_Customer'].fillna("NO", inplace=True)
final_df["C_TAT"].fillna(00, inplace=True)
final_df["V_TAT"].fillna(00, inplace=True)

```

Change datatypes & indtify problems with column vise

```python
final_df["Customer_Rating"] = final_df["Customer_Rating"].astype(int)
final_df['Driver_Ratings'] = final_df["Driver_Ratings"].astype(int)
final_df["V_TAT"]=final_df["V_TAT"].astype(float)
final_df["C_TAT"]=final_df["C_TAT"].astype(float)
```

**Check for Outliers and deal with them**

# **What to Do With Outliers (As a Data Analyst)**

-  Situation            &                	Action  
- Data entry mistake	       -        Remove the outlier
- Real value, but too extreme -	Cap it (set max limit)
- It makes trend misleading	 - Remove or explain it in reports
- You must keep all values	- Mention clearly in your analysis

```python
q1 = final_df["Booking_Value"].quantile(0.25)
q2 = final_df["Booking_Value"].quantile(0.75)

IQR = q2 - q1

lower = q1 - 1.5 * IQR
upper = q2 + 1.5 * IQR

outliers = final_df[(final_df['Booking_Value'] < lower) | (final_df['Booking_Value'] > upper)]
```

```python
# costomer will perfer which vechile most

fav_vechileof_costumer = final_df['Vehicle_Type'].value_counts()
print(fav_vechileof_costumer)
```

```python
# finding loyal costumer which book maximum success ride ( want to reward them )

#Filter for successful (completed) bookings
succesfull_bookings = final_df['Booking_Status'] == "Success"

# Group by Customer_ID and count successful rides
loyal_customers = (
    succesfull_bookings.groupby(final_df['Customer_ID'])
    .size()
    .reset_index(name='Successful_Rides')
    .sort_values(by='Successful_Rides', ascending=False)
)

loyal_customers.head(3)
```

```python
#finding the vechicle which generate maximun revune

#group by booking value by vechicle type
mostrrvaune_genratedvecicle = final_df.groupby('Vehicle_Type')['Booking_Value'].sum().sort_values(ascending=False)
print(mostrrvaune_genratedvecicle)
```

```python
#which time we get maximum order '

final_df['Time'] = pd.to_datetime(final_df['Time'], format='%H:%M:%S', errors='coerce')

#Check how many still failed:
print("Unparsed rows:", final_df['Time'].isna().sum())

#Change datatype innto datetime format
final_df['Time'] = pd.to_datetime(final_df['Time'], errors='coerce')

#exact all hours from time (dt.hour geting us hour data )
final_df['Hour'] = final_df['Time'].dt.hour

#count number of order by per hours
number_oforder_perhour = final_df['Hour'].value_counts().sort_index()
print(number_oforder_perhour)

#Find the peak hour
peak_hour = number_oforder_perhour.idxmax()
peak_orders = number_oforder_perhour.max()
mostorder = (f"Most orders came at {peak_hour}:00 with {peak_orders} orders.")
```

```python
#high cancellation rate area

canceled_status = final_df['Booking_Status'].isin(['Canceled by Driver', 'Canceled by Customer'])

cancellation_counts = final_df[canceled_status].groupby(['Booking_ID' ,'Pickup_Location']).size().reset_index(name='Cancel_Counts')
print(cancellation_counts)

cancellation_counts.sort_values(by='Cancel_Counts', ascending=False)
```
